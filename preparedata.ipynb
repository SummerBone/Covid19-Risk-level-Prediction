```python
#setup
import pandas as pd
import numpy as np

var_list = pd.read_csv('list_of_variables.txt')
var_list.head()
```

```python
var_list.Group.unique()
```

```python
#average temp

var_select = var_list[var_list.Group == 'Climate'].iloc[:, 0].unique()[12:24].tolist()

```

```python
#print(var_list[var_list.Group == 'Demographics'].iloc[:, 0].unique())

#pick 

var_select.append('FIPS')
var_select.extend(('Density per square mile of land area - Population','Total_age0to17', 'Total_age85plusr','Total_age65plus' , 'Total_age18to64' ))
```

```python
#print(var_list[var_list.Group == 'Education'].iloc[:, 0].unique())
var_select.extend(("Percent of adults with a bachelor's degree or higher 2014-18", 'Less than a high school diploma 2014-18'))

```

```python
#print(var_list[var_list.Group == 'Employment and median household income'].iloc[:, 0].unique())
var_select.append("Unemployment_rate_2018")
var_select.append("Median_Household_Income_2018")
```

```python
#print(var_list[var_list.Group == 'Healthcare'].iloc[:, 0].unique())
var_select.extend(('Active Physicians per 100000 Population 2018 (AAMC)', 
                 'MD and DO Student Enrollment per 100000 Population AY 2018-2019 (AAMC)', 'Total nurse practitioners (2019)',
                  'ICU Beds')  )
```

```python
#print(var_list[var_list.Group == 'Transit Scores'].iloc[:, 0].unique())
var_select.append("transit_scores - population weighted averages aggregated from town/city level to county")
```

```python
var_select
```

```python
ga_counties  = pd.read_csv('counties.csv')

ga_counties = ga_counties[var_select]

ga_counties.head()
```

```python
metro = pd.read_csv('Georgia_metro.csv')

metro['fips code'] = metro['fips code'].apply(lambda x:x[:5] )

metro['fips code'] = metro['fips code'].apply(lambda x:int(x) )

ga_counties = ga_counties.merge(metro, right_on='fips code', left_on='FIPS')

```

```python


#cases = pd.read_csv('georgia_case_death.csv')

cases = pd.read_csv('GA07_25.csv') #updated GA cases
cases = cases[cases['fips'].notna()]
cases = cases.drop(columns=['county','state'])
cases['fips'] = cases['fips'].apply(lambda x:int(x))  

```

```python
final = cases.merge(ga_counties, left_on = 'fips', right_on='FIPS')
final.head()
```

```python


cases_median = final["cases"].median()
deaths_median = final["deaths"].median()

def highrisk_cases(x):
    if x > cases_median:
        return 1
   
    else: return 0
    
def highrisk_deaths(x):
    if x > deaths_median:
        return 1
   
    else: return 0  
```

```python
final['High_risk_by_deaths'] = final["deaths"].apply(highrisk_deaths)
final['High_risk_by_cases'] = final["cases"].apply(highrisk_cases)
```

```python
final.to_csv('ga_county_covid.csv')
```
